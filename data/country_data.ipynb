```python
import pandas as pd
import numpy as np
```

```python
borders = pd.read_csv("https://raw.githubusercontent.com/geodatasource/country-borders/master/GEODATASOURCE-COUNTRY-BORDERS.CSV")
```

```python
borders = borders[['country_code','country_name','country_border_name']]
```

```python
borders_list = borders.groupby(['country_code', 'country_name']).agg(list).reset_index()
```

```python
borders_list.head()
```

```python
iso = pd.read_csv("https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv")
```

```python
iso.head()
```

```python
iso_sub =iso[['alpha-2', 'alpha-3']]
```

```python
borders_list = borders_list.merge(iso_sub, left_on='country_code', right_on='alpha-2', how="left")
```

```python
borders_list.head()
```

```python
borders_list = borders_list.mask(borders_list.applymap(str).eq('[nan]'))
```

```python
country_data = pd.read_csv("country_data.csv")
```

```python
country_data.head()
```

```python
full_data = country_data.merge(borders_list, left_on="iso_code", right_on="alpha-3", how="left")
```

```python
full_data = full_data.drop(['country_code', 'country_name','alpha-2', 'alpha-3'], axis=1)
```

```python
full_data.to_csv("full_country_data.csv")
```
